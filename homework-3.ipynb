```python
import matplotlib.pyplot as plt
import math

mu=0.5

logdens=[-15+0.01*i for i in range(2500)]

logT1=[logdens[i]/3+math.log10(3.2*mu**(-1/3))+7 for i in range(2500)]
logT2=[logdens[i]*2/3+math.log10(1.21*mu)+5 for i in range(2500)]
logT3=[logdens[i]/3+math.log10(1.5*mu)+7 for i in range(2500)]
logT4=[logdens[i]/3+math.log10(1.338)+3 for i in range(2500)]

cut1=int((math.log10(1.338)-math.log10(1.21*mu)-2)*300)+1500
cut2=int((math.log10(1.5*mu)-math.log10(1.21*mu)+2)*300)+1500

logx=[logdens[cut2] for i in range(100)]
logy=[(logT3[cut2]+6)/100*i-6 for i in range(100)]

for i in range(2500-cut1):
    logT4[i+cut1]='nan'

for i in range(2500-cut2):
    logT2[i+cut2]='nan'
    
for i in range(cut2):
    logT3[i]='nan'

plt.figure(figsize=(6,7.5))
plt.plot(logdens,logT1,'k--')
plt.plot(logdens,logT2,'k--')
plt.plot(logdens,logT3,'k--')
plt.plot(logdens,logT4,'k--')
plt.plot(logx,logy,'k--')
plt.xlabel('log$œÅ$($g/cm^{-3}$)',fontsize=14)
plt.ylabel('logT(K)',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('State Regimes\n',fontsize=16)
plt.text(-10,8,'I',fontsize=12)
plt.text(-8,3,'II',fontsize=12)
plt.text(-13,-3,'III',fontsize=12)
plt.text(0,0,'IV',fontsize=12)
plt.text(8,2,'V',fontsize=12)

plt.savefig('state_regimes.png')
plt.show()
```
